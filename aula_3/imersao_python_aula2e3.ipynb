```python
#importando os pacotes que utilizo no codigo
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
#como é uma planilha do excel com diversas abas, cada uma é carregada separadamente como um data frame independente
#carregando a principal

df_principal = pd.read_excel("Imersão Python - Tabela de ações.xlsx", sheet_name="Principal")
df_principal
```

```python
#carregando a quantidade de ações que foi movimentada no dia
df_total_acoes = pd.read_excel("Imersão Python - Tabela de ações.xlsx", sheet_name="Total_de_acoes")
df_total_acoes.head()
```

```python
#carregando o nome das empresas
df_ticker = pd.read_excel("Imersão Python - Tabela de ações.xlsx", sheet_name="Ticker")
df_ticker.head()
```

```python
#carregando segmento e idade
df_copilot = pd.read_excel("Imersão Python - Tabela de ações.xlsx", sheet_name="copilot")
df_copilot.head()
```

## Criando as novas colunas para análise e já corrigindo algumas questoes nos DFs

```python
#filtrando a tabela com apenas as colunas que serão trabalhadas
#o .copy() apaga o df original e cria um novo
#, "Val. Mín",	"Val. Máx",	"Volume"
df_principal = df_principal[["Ativo", "Data",	"Último (R$)",	"Var. Dia (%)"]].copy()

#renomeando as colunas para excluir espaços e caracteres especiais, que podem geral conflito no código
df_principal = df_principal.rename(columns={"Último (R$)": "valor_final", "Var. Dia (%)" : "var_dia_pct"}).copy()
df_principal
```

```python
#criando duas novas colunas:

#criando a coluna com a porcentagem em valor numérico decimal
df_principal['var_pct'] = df_principal['var_dia_pct']/100
#criando a coluna com o valor inicial da ação no dia
df_principal['valor_inicial'] = df_principal['valor_final']/(df_principal['var_pct']+1)
df_principal
```

```python
#puxando colunas das outras abas da planilha inicial (que agora são dfs separados)
df_principal = df_principal.merge(df_total_acoes, left_on= 'Ativo', right_on='Código', how = 'left')

#removendo o codigo das empresas pois é informação duplicada
df_principal = df_principal.drop(columns = ['Código'])
df_principal
```

```python
#renomeando a nova coluna
df_principal = df_principal.rename(columns={"Qtde. Teórica": "quantidade"})
df_principal
```

```python
#criando uma coluna para mostrar a totalidade em reais de movimentações para cada empresa,
#multiplicando a variação de uma unica ação pela quantidade de ações que foram movimentadas

df_principal['variacao_reais'] = (df_principal['valor_final'] - df_principal['valor_inicial'])*df_principal['quantidade']

#formatar os floats para exibição sem notação científica
pd.options.display.float_format = '{:.2f}'.format

df_principal
```

```python
df_principal
#a quantidade está como float, sendo que na realidade é um int
```

```python
#mudando a categoria da quantidade

df_principal['quantidade'] = df_principal['quantidade'].astype(int)
df_principal
```

```python
#criando a coluna com o resultado sobre queda ou subida (com auxilio do chat gpt)
#ele sugeriu utilizar a fç 'where' do numpy

# Aplicando a condição usando numpy.where
df_principal['resultado'] = np.where(df_principal['variacao_reais'] > 0, 'subiu',
                         np.where(df_principal['variacao_reais'] < 0, 'desceu', 'estavel'))
df_principal
```

```python
# puxando de outra tabela a coluna com o nome das empresas

df_principal = df_principal.merge(df_ticker, left_on= 'Ativo', right_on='Ticker', how = 'left')

df_principal = df_principal.drop(columns = ['Ticker'])
df_principal
```

```python
# agora puxando as informações do copilot
df_principal = df_principal.merge(df_copilot, left_on= 'Nome', right_on='Nome da empresa', how = 'left')

df_principal = df_principal.drop(columns = ['Nome da empresa'])
df_principal
```

## corrigindo as colunas de idade

```python
#agora eu vou limpar a bagunça que deixei de propósito

#vou checar a coluna que deu problema quando tentei classificar no excel
df_principal['idade']

#está como objeto
```

```python
#checando a coluna que consegui usar
df_principal['idade.1']
#aqui também saiu como objeto
```

```python
#quero retirar aqui os colchetes da coluna Idade (anos)

df_principal['Idade (anos)'] = df_principal['Idade (anos)'].apply(lambda x: str(x).replace("[", "").replace("]", ""))
df_principal
```

```python
df_principal['Idade (anos)'] = df_principal['Idade (anos)'].astype(int)
df_principal
```

```python
#parece que o erro de conversão da coluna idade foi pela existencia de valores nulos
#então quero ver qual o tipo de variavel de cada coluna e a quantidade de valores nulos
df_principal.info()
```

```python
#tirando uma das colunas de idade
df_principal = df_principal.drop(columns = ['idade'])
df_principal
```

```python
#renomeando a terceira coluna de idade. No excel fiquei com 3 colunas pois
# a primeira foi a resposta do copilot que colei diretamente na tabela
# na segunda eu fiz uma formula para retirar os colchetes, so que quando fui classificar as idades sairam todas com mais de 100 anos
# então eu fiz a terceira coluna pedindo para o chat GPT retirar os colchetes da primeira, e com essa consegui trabalhar

df_principal = df_principal.rename(columns={"idade.1": "idade"})
df_principal
```

```python
df_principal.isnull().sum()
```

```python
#Procurando a idade de empresa com valor nulo
valores_nulos = df_principal[df_principal['idade'].isnull()]
valores_nulos
```

```python
#pesquisando e colocando o valor faltante
#segmento saúde e 47 anos

df_principal['Segmento'][10] = 'Saude'
df_principal['Idade (anos)'][10] = 47
df_principal['idade'][10] = 47
```

```python
#conferindo se os valores foram substituidos
df_principal.info()
```

```python
#conferindo se os valores foram substituidos corretamente na linha que estava errada
df_principal.iloc[10]
```

```python
df_principal.drop(columns = ['Idade (anos)'], inplace=True)
```

```python
df_principal['idade'] = df_principal['idade'].astype(int)
df_principal
```

```python
df_principal[df_principal['idade'].str.contains('-   ')]
#o str contains não retorna valor booleano, por isso não funciona aqui
```

```python
#o chat GPT sugeriu fazer uma função que retornasse booleano
def verificar_problema(valor):
    if valor == '-':
        return True
    else:
        return False

# Aplicando a função na coluna 'idade' para identificar a linha problemática
df_principal[df_principal['idade'].apply(verificar_problema)]

#porém também não funcionou, sendo esta a teoria do GPT:
#  "Parece que o problema persiste porque o pandas trata os valores problemáticos
#   como NaN, e isso está causando o erro ao tentar usar .str.contains() ou .apply()."
```

```python
#Nova sugestão de correção do erro
# Substituindo os valores problemáticos por NaN
df_principal['idade'] = df_principal['idade'].replace('-   ', np.nan)

# Identificando a linha onde o valor problemático foi substituído por NaN
df_principal[df_principal['idade'].isnull()]
```

```python
# novamente corrigirei manualmente pesquisando sobre a empresa,
# bem como corrigirei o nome dela que continua aparecendo como código da bolsa
#esta empresa é a Aliansce Sonae Shopping Centers
# o segmento dela é Shopping Centers
# ela tem 21 anos

df_principal['Nome'][55] = 'Aliansce Sonae Shopping Centers'
df_principal['Segmento'][55] = 'Shopping Centers'
df_principal['idade'][55] = 21
df_principal.iloc[55]
```

```python
# Convertendo a coluna 'idade' para o tipo int
df_principal['idade'] = df_principal['idade'].astype(int)
df_principal.info()
```

```python
#agora dividindo a idade em categorias

df_principal['cat_idade'] = np.where(df_principal['idade'] < 50, 'menos de 50',
                        np.where(df_principal['idade'] > 100, 'mais de 100', 'entre 50 e 100'))
df_principal.head()
```

```python
#exportando a nova tabela limpa e reorganizada
# Exportar o DataFrame para um arquivo CSV
df_principal.to_csv('dados.csv', index=False)  # index=False para não incluir o índice do DataFrame no arquivo CSV
```

```python
# terceiro desafio: Fazer análise por faixa etária gerando soma, contagem e tambem um gráfico de barras

#groupby para soma e contagem
#usando a fç agg:
#'usada para aplicar uma ou mais funções de agregação a um DataFrame ou a um grupo de dados gerado por groupby' -GPT, chat

df_faixa_etaria = df_principal.groupby('cat_idade')['variacao_reais'].agg(['sum', 'count']).reset_index()
df_faixa_etaria
```

```python
#agora o grafico de barras

fig = px.bar(df_faixa_etaria, x='cat_idade', y='sum', text='sum',
             title = 'variação total em reais por faixa de idade')
#fig.update_layout(yaxis_tickformat=',.2f', yaxis_title='Variação em Reais')
fig.update_traces(texttemplate='%{text:,.2f}', textposition='outside', hovertemplate='%{y:,.2f} Reais')
fig.show()
```

```python
# quero colocar também a contagem de empresas em cada categoria
fig = px.bar(df_faixa_etaria, x='cat_idade', y='sum', text='count',
             labels={'cat_idade': 'Categoria de Idade', 'sum': 'soma'},
             title='variação total em reais por faixa de idade')

# Adicionar as contagens no topo das barras
fig.update_traces(texttemplate='%{text}', textposition='outside')
fig.update_traces(texttemplate='%{text:,.2f}', textposition='outside', hovertemplate='%{y:,.2f} Reais')

#salvando o png dessa figura
fig.write_image('grafico_barras.png')
fig.show()
```
